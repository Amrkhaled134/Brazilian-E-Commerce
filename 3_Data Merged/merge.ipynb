```python
df_orders=pd.read_csv(r"C:\Users\SAMA\Documents\Brazilian E-Commerce\2_Cleaned Data\orders_clean.csv")
df_products=pd.read_csv(r"C:\Users\SAMA\Documents\Brazilian E-Commerce\2_Cleaned Data\products_clean.csv")
df_order_items=pd.read_csv(r"C:\Users\SAMA\Documents\Brazilian E-Commerce\2_Cleaned Data\order_items_clean.csv")
df_payment=pd.read_csv(r"C:\Users\SAMA\Documents\Brazilian E-Commerce\2_Cleaned Data\payment_clean.csv")
df_customers=pd.read_csv(r"C:\Users\SAMA\Documents\Brazilian E-Commerce\2_Cleaned Data\customers_clean.csv")
df_reviews=pd.read_csv(r"C:\Users\SAMA\Documents\Brazilian E-Commerce\2_Cleaned Data\reviews_clean.csv")
df_sellers=pd.read_csv(r"C:\Users\SAMA\Documents\Brazilian E-Commerce\2_Cleaned Data\sellers_clean.csv")
```
